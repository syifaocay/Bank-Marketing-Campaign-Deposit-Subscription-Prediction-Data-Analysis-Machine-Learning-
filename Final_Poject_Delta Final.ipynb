## Background

Untuk memasarkan produk ini, bank sering menggunakan strategi direct marketing, terutama melalui panggilan telepon langsung kepada nasabah.<br>

## Modeling Approach & Evaluation Metrics

Analysis ini menggunakan pendekatan klasifikasi untuk memprediksi kemungkinan nasabah berlangganan deposito berjangka (target: *yes* atau *no*). Tahap awal meliputi preprocessing data seperti penanganan missing values, encoding variabel kategorikal, serta pembagian data menjadi data latih dan data uji menggunakan *train-test split* dengan *stratified sampling* untuk menjaga proporsi kelas. Pada kondisi data tidak seimbang, dapat diterapkan teknik oversampling untuk meningkatkan kemampuan model dalam mendeteksi kelas minoritas.<br><br>

Beberapa algoritma klasifikasi yang digunakan antara lain Logistic Regression, Decision Tree, Random Forest, K-Nearest Neighbors (KNN), XGBoost, dan LightGBM. Penggunaan beberapa model bertujuan untuk membandingkan performa dan memilih model terbaik. Selain itu, diterapkan *cross-validation* untuk mengurangi risiko overfitting dan memastikan stabilitas performa model.<br><br>

Saat ini, proses penentuan calon nasabah yang akan dihubungi masih belum optimal, sehingga berpotensi menimbulkan dua jenis kesalahan prediksi:
1. False Negative (FN)
Model memprediksi calon nasabah tidak tertarik sehingga tidak dihubungi, padahal sebenarnya nasabah tersebut memiliki potensi untuk membuka akun.
Dampaknya:
- Kehilangan peluang penjualan
- Potensi revenue tidak terealisasi
- Opportunity cost meningkat

2. False Positive (FP)
Model memprediksi calon nasabah tertarik sehingga dilakukan panggilan, namun ternyata nasabah tersebut tidak berminat.
Dampaknya:
- Pemborosan biaya telemarketing
- Waktu agen terbuang

Karena kedua jenis kesalahan tersebut sama-sama memberikan dampak finansial dan operasional yang signifikan, maka diperlukan model klasifikasi yang mampu menjaga keseimbangan antara precision dan recall. <br><br>

Untuk mengatasi permasalahan tersebut, model dikembangkan dengan fokus pada optimasi F1-score, karena metrik ini merupakan mean antara precision dan recall.
Dengan menggunakan F1-score sebagai metrik utama, model diharapkan dapat:
- Meminimalkan kehilangan peluang nasabah potensial (menekan FN)
- Mengurangi pemborosan biaya panggilan yang tidak efektif (menekan FP)
- Menghasilkan strategi targeting yang lebih efisien dan seimbang

Dengan pendekatan ini, diharapkan diperoleh model yang tidak hanya memiliki performa statistik yang baik, tetapi juga mampu mendukung strategi penargetan dan optimasi kampanye pemasaran secara efektif.

### HISTOGRAM DATA NUMERIK KONTINU

```python

colors_hist = sns.color_palette("Set2", len(continuous_columns))

# HISTOGRAM UNTUK DATA KONTINU (VERSI OPTIMAL)
fig1, axes1 = plt.subplots(1, 3, figsize=(20, 7))

for idx, col in enumerate(continuous_columns):
    # Data preprocessing singkat untuk handle outliers visual (opsional namun disarankan)
    data = bank[col].dropna()
    
    # Histogram dengan density=True agar KDE bisa menempel dengan pas
    # Menggunakan warna dari palette Set2 dengan alpha 0.8 dan edgecolor black
    axes1[idx].hist(data, bins=40, color=colors_hist[idx],
                    alpha=0.8, edgecolor='black', linewidth=1, density=True)

    # Tambahkan KDE curve (garis merah gelap untuk kontras)
    sns.kdeplot(data, ax=axes1[idx], color='darkred', linewidth=2.5)

    # Statistik Dasar
    mean_val = data.mean()
    median_val = data.median()

    # Garis Mean & Median (Gaya garis putus-putus sesuai permintaan sebelumnya)
    axes1[idx].axvline(mean_val, color='red', linestyle='--', 
                       linewidth=2, label=f'Mean: {mean_val:.2f}')
    axes1[idx].axvline(median_val, color='blue', linestyle='--', 
                       linewidth=2, label=f'Median: {median_val:.2f}')

    # Labels & Styling
    axes1[idx].set_xlabel(col.upper(), fontsize=12, fontweight='bold')
    axes1[idx].set_ylabel('Density', fontsize=11)
    
    # Judul dengan tambahan informasi ringkas di atasnya
    axes1[idx].set_title(f'Distribusi {col.upper()} (Kontinu)\n' +
                         f'Mean={mean_val:.2f}, Median={median_val:.2f}',
                         fontsize=13, fontweight='bold', pad=15)
    
    axes1[idx].legend(loc='upper right', frameon=True, fontsize=10)
    axes1[idx].grid(axis='y', alpha=0.3)

    # Mengatur limit sumbu X agar tidak terlalu lebar karena outliers (meningkatkan keterbacaan)
    # Menampilkan data dari persentil 0.1 ke 99.9
    axes1[idx].set_xlim(data.min(), data.quantile(0.999))

plt.tight_layout()
plt.show()
```

## Handling Missing Value

Dataset ini tidak mengandung missing value secara eksplisit, namun beberapa fitur kategorikal memiliki nilai 'unknown' yang perlu ditangani agar tidak menimbulkan bias pada model:



- **job**  
  Nilai `unknown` diisi dengan kategori **'other'**, karena:  
  - Jumlahnya relatif kecil  
  - Tidak merepresentasikan jenis pekerjaan yang spesifik  
  → Penggabungan ke kategori lain lebih aman daripada dibiarkan ambigu.

- **marital**  
  Nilai `unknown` diimputasi menggunakan **modus** (yaitu: **married**), karena:  
  - Jumlahnya sangat sedikit  
  → Penggantian dengan nilai paling umum tidak akan mengubah distribusi secara berarti.

- **education**  
  Nilai `unknown` **dibiarkan sebagai kategori tersendiri**, karena:  
  - Jumlahnya cukup signifikan  
  - Imputasi dengan modus atau penghapusan berisiko menghilangkan informasi penting dan mengubah distribusi fitur secara tidak proporsional.

- **default**  
  Nilai `unknown` **dibiarkan sebagai kategori tersendiri**, karena:  
  - Kemungkinan besar mencerminkan kondisi nyata (status kredit nasabah tidak dapat diverifikasi)  
  - Bukan sekadar missing value acak.

- **housing** & **loan**  
  Nilai `unknown` **dibiarkan sebagai kategori tersendiri** (sama seperti default), karena:  
  - Mengindikasikan ketidaklengkapan informasi yang tidak bisa diasumsikan secara sembarangan  
  - Imputasi berisiko menyebabkan salah interpretasi.

- **poutcome**  
  Nilai **'nonexistent'** **dibiarkan sebagai kategori tersendiri**, karena:  
  - Merepresentasikan nasabah yang **belum pernah dihubungi sebelumnya**  
  - Segmen ini berpotensi memiliki makna tersendiri bagi model prediksi.





```python
# Handling job: unknown → 'other'
df['job'] = df['job'].replace('unknown', 'other')

# Handling marital: unknown → modus
modus_marital = df['marital'].mode()[0]
df['marital'] = df['marital'].replace('unknown', modus_marital)

# cek fitur kolom job dan marital
print(df['job'].value_counts(dropna=False))
print()
print(df['marital'].value_counts(dropna=False))
```

## Model Benchmarking dengan Kombinasi Imbalance Handling dan Cross-Validation

```python
# ======================
# --- Definisi Model ---
# ======================
models_no_weight = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree':       DecisionTreeClassifier(random_state=42),
    'Random Forest':       RandomForestClassifier(random_state=42),
    'KNN':                 KNeighborsClassifier(),
    'XGBoost':             XGBClassifier(random_state=42),
    'LightGBM':            LGBMClassifier(random_state=42)
}

# Model dengan built-in class_weight — HANYA dipakai dengan resample='none'
models_with_weight = {
    'Logistic Regression': LogisticRegression(random_state=42, class_weight='balanced'),
    'Decision Tree':       DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'Random Forest':       RandomForestClassifier(random_state=42, class_weight='balanced'),
    'KNN':                 KNeighborsClassifier(),                                    # tidak support class_weight
    'XGBoost':             XGBClassifier(random_state=42, scale_pos_weight=8),
    'LightGBM':            LGBMClassifier(random_state=42, class_weight='balanced')
}

# ============================
# --- Imbalance Treatment ---
# ============================

resamples = {
    'none':        None,
    'ros':         RandomOverSampler(random_state=20),
    'rus':         RandomUnderSampler(random_state=20),
    'smote':       SMOTE(random_state=20),
    'ncr':         NeighbourhoodCleaningRule(),
    'smote_tomek': SMOTETomek(random_state=20)
}

# ============================
# --- Metrics ---
# ============================

metrics = {
    'pr_auc':    'average_precision',
    'precision': make_scorer(precision_score, pos_label=1, zero_division=0),
    'recall':    make_scorer(recall_score,    pos_label=1, zero_division=0),
    'f1':        make_scorer(f1_score,        pos_label=1, zero_division=0)
}

# ============================
# --- Container Result ---
# ============================

test_pr_auc_arr    = []
test_precision_arr = []
test_recall_arr    = []
test_f1_arr        = []
list_model         = []
list_resample      = []

# ============================
# --- Cross Validation Loop ---
# ============================

for name_resample, resample in resamples.items():
    for name_model in models_no_weight.keys():

        # ==============================
        # PILIH MODEL — hindari double handling
        # ==============================

        if resample is None:
            # Tidak ada resampler → gunakan class_weight sebagai satu-satunya penanganan
            model = models_with_weight[name_model]
        else:
            # Ada resampler → gunakan model TANPA class_weight
            model = models_no_weight[name_model]

        crossval = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

        # ==============================
        # PILIH PREPROCESSOR
        # ==============================

        if name_model in ['Logistic Regression', 'KNN']:
            chosen_preprocessor = preprocessor_scaled
        else:
            chosen_preprocessor = preprocessor_noscale

        # ==============================
        # BUILD PIPELINE
        # ==============================

        if resample is None:
            estimator = Pipeline([
                ('preprocessor', chosen_preprocessor),
                ('model', model),
            ])
        else:
            estimator = Pipeline([
                ('preprocessor', chosen_preprocessor),
                ('resample', resample),
                ('model', model),
            ])

        list_model.append(name_model)
        list_resample.append(name_resample)

        scores = cross_validate(
            estimator,
            X_train,
            y_train,
            cv=crossval,
            scoring=metrics,
            n_jobs=-1
        )

        test_pr_auc_arr.append(scores['test_pr_auc'].mean() * 100)
        test_precision_arr.append(scores['test_precision'].mean() * 100)
        test_recall_arr.append(scores['test_recall'].mean() * 100)
        test_f1_arr.append(scores['test_f1'].mean() * 100)

# ============================
# --- Benchmark Result ---
# ============================

model_bench = pd.DataFrame({
    'model':     list_model,
    'resample':  list_resample,
    'F1':        test_f1_arr,
    'PR_AUC':    test_pr_auc_arr,
    'recall':    test_recall_arr,
    'precision': test_precision_arr
    
})

print("\n" + "="*80)
print("MODEL BENCHMARKING RESULTS (Sorted by F1 Score)")
print("="*80)
print(model_bench.sort_values(by='F1', ascending=False))
```

## Hyperparameter Tuning XGBoost dengan NCR Menggunakan GridSearchCV

```python
# ============================
# --- Setup Model & NCR ---
# ============================

from xgboost import XGBClassifier

xgb = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False,
    verbosity=0
)

ncr = NeighbourhoodCleaningRule()

# ============================
# --- Pipeline ---
# ============================

pipeline = Pipeline([
    ('preprocessor', preprocessor_noscale),
    ('resampling', ncr),
    ('model', xgb)
])

# ============================
# --- Parameter Grid ---
# ============================

param_grid = {
    'model__n_estimators':      [100, 300],
    'model__learning_rate':     [0.05, 0.01],
    'model__max_depth':         [3, 5, 10],
    'model__min_child_weight':  [1, 5, 10],
    'model__subsample':         [0.8, 1.0],
    'model__colsample_bytree':  [0.8, 1.0]
}

print("="*80)
print("STARTING HYPERPARAMETER TUNING - XGBoost with NCR")
print("="*80)

total_combination = (
    len(param_grid['model__n_estimators']) *
    len(param_grid['model__learning_rate']) *
    len(param_grid['model__max_depth']) *
    len(param_grid['model__min_child_weight']) *
    len(param_grid['model__subsample']) *
    len(param_grid['model__colsample_bytree'])
)

print(f"Total combinations to try: {total_combination}")

# ============================
# --- GridSearchCV (Multi Scoring) ---
# ============================

scoring = {
    'pr_auc':    'average_precision',
    'recall':    'recall',
    'precision': 'precision',
    'f1':        'f1'
}

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring=scoring,
    refit='f1',
    n_jobs=-1,
    verbose=2,
    return_train_score=True
)

# ============================
# --- Fit ---
# ============================

grid_search.fit(X_train, y_train)
```

```python
# Encode target
y_train = y_train.map({'no': 0, 'yes': 1})
y_test  = y_test.map({'no': 0, 'yes': 1})


# ======================
# 1. NUMERICAL COLUMNS
# ======================
num_cols = [
    'age',
    'previous',
    'cons.price.idx',
    'cons.conf.idx',
    'euribor3m',
    'risk_score',
    'month_num',
    'was_contacted_before',
    'nr.employed',
    'emp.var.rate'
]

# ======================
# 2. ONE HOT COLUMNS
# ======================
cat_cols = [
    'job',
    'marital',
    'day_of_week',
    'default',
    'housing',
    'contact',
    'poutcome',
    'loan'
]

# ======================
# 3. ORDINAL COLUMNS
# ======================
ord_cols = ['education', 'age_group', 'campaign_group']

education_order = [[
    'unknown',
    'illiterate',
    'basic.4y',
    'basic.6y',
    'basic.9y',
    'high.school',
    'professional.course',
    'university.degree'
]]

age_group_order = [[
    'young',
    'adult',
    'middle_age',
    'senior'
]]

campaign_group_order = [[
    'Very Low',
    'Low',
    'Medium',
    'High',
    'Very High'
]]

ordinal_categories = education_order + age_group_order + campaign_group_order

# ======================
# 4. COLUMN TRANSFORMER
# ======================
preprocessor_scaled = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), cat_cols),
        ('ordinal', OrdinalEncoder(categories=ordinal_categories), ord_cols)
    ],
    remainder='drop'
)

# ======================
# WITHOUT SCALING
# ======================

preprocessor_noscale = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', num_cols),  # TANPA scaling
        ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), cat_cols),
        ('ordinal', OrdinalEncoder(categories=ordinal_categories), ord_cols)
    ],
    remainder='drop'
)
```

```python
# ======================
# --- Definisi Model ---
# ======================
models_no_weight = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree':       DecisionTreeClassifier(random_state=42),
    'Random Forest':       RandomForestClassifier(random_state=42),
    'KNN':                 KNeighborsClassifier(),
    'XGBoost':             XGBClassifier(random_state=42),
    'LightGBM':            LGBMClassifier(random_state=42)
}

# Model dengan built-in class_weight — HANYA dipakai dengan resample='none'
models_with_weight = {
    'Logistic Regression': LogisticRegression(random_state=42, class_weight='balanced'),
    'Decision Tree':       DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'Random Forest':       RandomForestClassifier(random_state=42, class_weight='balanced'),
    'KNN':                 KNeighborsClassifier(),                                    # tidak support class_weight
    'XGBoost':             XGBClassifier(random_state=42, scale_pos_weight=8),
    'LightGBM':            LGBMClassifier(random_state=42, class_weight='balanced')
}

# ============================
# --- Imbalance Treatment ---
# ============================

resamples = {
    'none':        None,
    'ros':         RandomOverSampler(random_state=20),
    'rus':         RandomUnderSampler(random_state=20),
    'smote':       SMOTE(random_state=20),
    'ncr':         NeighbourhoodCleaningRule(),
    'smote_tomek': SMOTETomek(random_state=20)
}

# ============================
# --- Metrics ---
# ============================

metrics = {
    'pr_auc':    'average_precision',
    'precision': make_scorer(precision_score, pos_label=1, zero_division=0),
    'recall':    make_scorer(recall_score,    pos_label=1, zero_division=0),
    'f1':        make_scorer(f1_score,        pos_label=1, zero_division=0)
}

# ============================
# --- Container Result ---
# ============================

test_pr_auc_arr    = []
test_precision_arr = []
test_recall_arr    = []
test_f1_arr        = []
list_model         = []
list_resample      = []

# ============================
# --- Cross Validation Loop ---
# ============================

for name_resample, resample in resamples.items():
    for name_model in models_no_weight.keys():

        # ==============================
        # PILIH MODEL — hindari double handling
        # ==============================

        if resample is None:
            # Tidak ada resampler → gunakan class_weight sebagai satu-satunya penanganan
            model = models_with_weight[name_model]
        else:
            # Ada resampler → gunakan model TANPA class_weight
            model = models_no_weight[name_model]

        crossval = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

        # ==============================
        # PILIH PREPROCESSOR
        # ==============================

        if name_model in ['Logistic Regression', 'KNN']:
            chosen_preprocessor = preprocessor_scaled
        else:
            chosen_preprocessor = preprocessor_noscale

        # ==============================
        # BUILD PIPELINE
        # ==============================

        if resample is None:
            estimator = Pipeline([
                ('preprocessor', chosen_preprocessor),
                ('model', model),
            ])
        else:
            estimator = Pipeline([
                ('preprocessor', chosen_preprocessor),
                ('resample', resample),
                ('model', model),
            ])

        list_model.append(name_model)
        list_resample.append(name_resample)

        scores = cross_validate(
            estimator,
            X_train,
            y_train,
            cv=crossval,
            scoring=metrics,
            n_jobs=-1
        )

        test_pr_auc_arr.append(scores['test_pr_auc'].mean() * 100)
        test_precision_arr.append(scores['test_precision'].mean() * 100)
        test_recall_arr.append(scores['test_recall'].mean() * 100)
        test_f1_arr.append(scores['test_f1'].mean() * 100)

# ============================
# --- Benchmark Result ---
# ============================

model_bench = pd.DataFrame({
    'model':     list_model,
    'resample':  list_resample,
    'F1':        test_f1_arr,
    'PR_AUC':    test_pr_auc_arr,
    'recall':    test_recall_arr,
    'precision': test_precision_arr
    
})

print("\n" + "="*80)
print("MODEL BENCHMARKING RESULTS (Sorted by F1 Score)")
print("="*80)
print(model_bench.sort_values(by='F1', ascending=False))
```

```python
# ============================
# --- Ambil F1 terbaik per model ---
# ============================

best_per_model = (
    model_bench
    .loc[model_bench.groupby('model')['F1'].idxmax()]
)

# ============================
# --- Ambil Top 3 Final ---
# ============================

top3_unique_models = (
    best_per_model
    .sort_values(by='F1', ascending=False)
    .head(3)
    .reset_index(drop=True)
)

print("\n" + "="*80)
print("TOP 3 UNIQUE MODELS (F1 Highest)")
print("="*80)
print(top3_unique_models)
```

```python
# ======================================================
# --- MODEL DEFINITIONS ---
# ======================================================

models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost':       XGBClassifier(random_state=42),
    'LightGBM':      LGBMClassifier(random_state=42)
}

# ======================================================
# --- RESAMPLING ---
# ======================================================

selected_resamples = {
    'NCR': NeighbourhoodCleaningRule()
}

metrics_list     = []
conf_matrix_list = []

# ======================================================
# --- EVALUATION LOOP ---
# ======================================================

for name_resample, resample in selected_resamples.items():
    for name_model, model in models.items():

        combo_name = f"{name_model} ({name_resample})"

        # ==========================
        # SELECT PREPROCESSOR
        # ==========================

        chosen_preprocessor = preprocessor_noscale  # RF, XGB, LGBM tidak butuh scaling

        # ==========================
        # BUILD PIPELINE
        # ==========================

        pipeline = Pipeline([
            ('preprocessor', chosen_preprocessor),
            ('resampling',   resample),
            ('model',        model)
        ])

        # ==========================
        # FIT TRAIN
        # ==========================

        pipeline.fit(X_train, y_train)

        # ==========================
        # PREDICT TEST
        # ==========================

        y_pred  = pipeline.predict(X_test)
        y_proba = pipeline.predict_proba(X_test)[:, 1]

        pr_auc = average_precision_score(y_test, y_proba)

        metrics_list.append([
            name_model,
            name_resample,
            combo_name,
            precision_score(y_test, y_pred, pos_label=1, zero_division=0) * 100,
            recall_score(y_test, y_pred,    pos_label=1, zero_division=0) * 100,
            f1_score(y_test, y_pred,        pos_label=1, zero_division=0) * 100,
            pr_auc * 100
        ])


# ======================================================
# --- RESULT DATAFRAME ---
# ======================================================

metrics_df = pd.DataFrame(
    metrics_list,
    columns=['Model', 'Resample', 'Resample', 'Precision', 'Recall', 'F1', 'PR_AUC']
)

print("\n" + "="*80)
print("EVALUATION ON TEST SET (Sorted by F1)")
print("="*80)
print(metrics_df.sort_values('F1', ascending=False).to_string(index=False))

# ======================================================
# --- BEST MODEL (F1 Tertinggi) ---
# ======================================================

best_model = metrics_df.sort_values('F1', ascending=False).iloc[0]

print("\n" + "="*80)
print("BEST MODEL (F1 Tertinggi)")
print("="*80)
print(best_model)
```

```python
# ============================
# --- Setup Model & NCR ---
# ============================

lgbm = LGBMClassifier(
    random_state=42,
    verbosity=-1
)

ncr = NeighbourhoodCleaningRule()

# ============================
# --- Pipeline ---
# ============================

pipeline = Pipeline([
    ('preprocessor', preprocessor_noscale),
    ('resampling', ncr),
    ('model', lgbm)
])

# ============================
# --- Parameter Grid ---
# ============================

param_grid = {
    
    'model__n_estimators':      [100, 300],
    'model__learning_rate':     [0.05, 0.01],
    'model__num_leaves':        [31, 64],
    'model__max_depth':         [-1, 5, 10],
    'model__min_child_samples': [20, 40]
}

print("="*80)
print("STARTING HYPERPARAMETER TUNING - LightGBM with NCR")
print("="*80)

total_combination = (
    len(param_grid['model__n_estimators']) *
    len(param_grid['model__learning_rate']) *
    len(param_grid['model__num_leaves']) *
    len(param_grid['model__max_depth']) *
    len(param_grid['model__min_child_samples'])
)

print(f"Total combinations to try: {total_combination}")

# ============================
# --- GridSearchCV (Multi Scoring) ---
# ============================

scoring = {
    'pr_auc':    'average_precision',
    'recall':    'recall',
    'precision': 'precision',
    'f1':        'f1'
}

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring=scoring,
    refit='f1',
    n_jobs=-1,
    verbose=2,
    return_train_score=True
)

# ============================
# --- Fit ---
# ============================

grid_search.fit(X_train, y_train)
```

```python
# ============================
# --- Ambil Best Model ---
# ============================
best_pipeline = grid_search.best_estimator_

# ============================
# --- Predict Test Set ---
# ============================
y_test_proba = best_pipeline.predict_proba(X_test)[:, 1]
y_test_pred  = best_pipeline.predict(X_test)

# ============================
# --- Calculate Metrics (TEST SET) ---
# ============================
test_pr_auc    = average_precision_score(y_test, y_test_proba) * 100
test_recall    = recall_score(y_test, y_test_pred, pos_label=1, zero_division=0) * 100
test_precision = precision_score(y_test, y_test_pred, pos_label=1, zero_division=0) * 100
test_f1        = f1_score(y_test, y_test_pred, pos_label=1, zero_division=0) * 100

# (Opsional) CV score tetap bisa ditampilkan sebagai referensi
best_index = grid_search.best_index_
cv_pr_auc    = grid_search.cv_results_['mean_test_pr_auc'][best_index] * 100
cv_recall    = grid_search.cv_results_['mean_test_recall'][best_index] * 100
cv_precision = grid_search.cv_results_['mean_test_precision'][best_index] * 100
cv_f1        = grid_search.cv_results_['mean_test_f1'][best_index] * 100

# ============================
# --- Display Results ---
# ============================
print("\n" + "="*80)
print("BEST PARAMETERS (LightGBM + NCR)")
print("="*80)
for param, value in grid_search.best_params_.items():
    print(f"{param}: {value}")

print("\n" + "-"*40)
print("SCORES TUNING (Train Validation)")
print("-"*40)
print(f"F1       : {cv_f1:.2f}%")
print(f"PR-AUC   : {cv_pr_auc:.2f}%")
print(f"Recall   : {cv_recall:.2f}%")
print(f"Precision: {cv_precision:.2f}%")


print("\n" + "-"*40)
print("TEST SET SCORES")
print("-"*40)
print(f"F1       : {test_f1:.2f}%")
print(f"PR-AUC   : {test_pr_auc:.2f}%")
print(f"Recall   : {test_recall:.2f}%")
print(f"Precision: {test_precision:.2f}%")
```
